```python
import pygame


def main():
    pygame.init()
    screen = pygame.display.set_mode((640, 480))
    clock = pygame.time.Clock()

    radius = 15
    color = (0, 0, 255)  # Начальный цвет - синий
    drawing = False
    shape = "line"  # Может быть "line", "rectangle", "circle", "eraser"
    start_pos = (0, 0)
    background_color = (0, 0, 0)  # Черный фон

    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                return

            if event.type == pygame.KEYDOWN:
                # Смена цвета
                if event.key == pygame.K_r:
                    color = (255, 0, 0)  # красный
                    shape = "line"
                elif event.key == pygame.K_g:
                    color = (0, 255, 0)  # зеленый
                    shape = "line"
                elif event.key == pygame.K_y:
                    color = (255, 255, 0)  # желтый
                    shape = "line"
                elif event.key == pygame.K_b:
                    color = (0, 0, 255)  # синий
                    shape = "line"

                # Смена инструмента
                elif event.key == pygame.K_t:
                    shape = "rectangle"
                elif event.key == pygame.K_c:
                    shape = "circle"
                elif event.key == pygame.K_e:
                    shape = "eraser"

                # Очистка экрана по пробелу
                elif event.key == pygame.K_SPACE:
                    screen.fill(background_color)

            if event.type == pygame.MOUSEBUTTONDOWN:
                if event.button == 1:  # левая кнопка мыши
                    drawing = True
                    start_pos = event.pos
                    if shape == "line":
                        pygame.draw.circle(screen, color, event.pos, radius)

            if event.type == pygame.MOUSEBUTTONUP:
                if event.button == 1:
                    drawing = False
                    if shape == "rectangle":
                        end_pos = event.pos
                        rect = pygame.Rect(start_pos, (end_pos[0] - start_pos[0], end_pos[1] - start_pos[1]))
                        pygame.draw.rect(screen, color, rect, radius)
                    elif shape == "circle":
                        end_pos = event.pos
                        dx = end_pos[0] - start_pos[0]
                        dy = end_pos[1] - start_pos[1]
                        circle_radius = max(abs(dx), abs(dy))
                        pygame.draw.circle(screen, color, start_pos, circle_radius, radius)

            if event.type == pygame.MOUSEMOTION:
                if drawing:
                    if shape == "line":
                        pygame.draw.circle(screen, color, event.pos, radius)
                    elif shape == "eraser":
                        pygame.draw.circle(screen, background_color, event.pos, radius)

        pygame.display.flip()
        clock.tick(60)


if __name__ == "__main__":
    main()
```
